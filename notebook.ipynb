```python
import re

def extract_signal_mappings_from_instances(verilog_file):
    """Extracts signal mappings from module instances in the synthesized netlist."""
    mappings = {}
    pattern = r"\.(\w+)\((\S+)\)"  # Matches instance connections like `.B1(_1684_)`

    with open(verilog_file, "r") as file:
        for line in file:
            match = re.findall(pattern, line)
            if match:
                for port, signal in match:
                    mappings[signal] = port  # Store mapping

    return mappings

# Run Extraction
verilog_file_path = "xtea.synthesis_preroute.v"
signal_mappings = extract_signal_mappings_from_instances(verilog_file_path)

# Save Mappings to CSV
import pandas as pd
df_mappings = pd.DataFrame(signal_mappings.items(), columns=["Synthesized Name", "Original Name"])
df_mappings.to_csv("signal_mappings.csv", index=False)

# Print first few mappings
print(df_mappings.head())
```

```python
import re
from collections import defaultdict
import pandas as pd

def extract_fan_in_out(netlist_file):
    """Extracts fan-in and fan-out counts from a Verilog netlist."""
    fan_in = defaultdict(int)  # Fan-in count: How many times a signal is used as an input
    fan_out = defaultdict(int)  # Fan-out count: How many times a signal is driving an output
    
    signal_usage = defaultdict(list)  # Track where each signal is used (for debugging)
    
    with open(netlist_file, "r") as file:
        lines = file.readlines()

    for line in lines:
        # Skip unnecessary cells (e.g., decap, filler, diode, power, ground)
        if any(x in line.lower() for x in ["decap", "fill", "diode", "tapvpwrvgnd", "vgnd", "vpwr"]):
            continue

        # Regex to find pin connections: pin names and associated signals
        matches = re.findall(r"\.(\w+)\((\S+)\)", line)

        if matches:
            outputs = []
            inputs = []
            for pin, signal in matches:
                if pin in {"Y", "Q", "ZN"}:  # Output pins (Y, Q, ZN, etc.)
                    outputs.append(signal)
                else:  # Input pins (A, B, A1, B1, etc.)
                    inputs.append(signal)
                    signal_usage[signal].append(outputs[0] if outputs else None)  # Track signal usage

            # Increment Fan-Out for each output signal (number of times it's used as input)
            for out_sig in outputs:
                fan_out[out_sig] += len(inputs)  # Output is driving multiple inputs

            # Increment Fan-In for each input signal
            for in_sig in inputs:
                fan_in[in_sig] += 1

    # Debug: Print first 10 usages of each signal
    print("\nSignal Usage (First 10):")
    for i, (sig, destinations) in enumerate(signal_usage.items()):
        if i >= 10:
            break
        print(f"{sig} is used in {destinations}")
    
    return fan_in, fan_out

# Run Extraction
netlist_file_path = "xtea.synthesis_preroute.v"  # Adjust the path as necessary
fan_in, fan_out = extract_fan_in_out(netlist_file_path)

# Convert to DataFrame and Save
df_fanin = pd.DataFrame(list(fan_in.items()), columns=["Signal", "Fan-in"])
df_fanout = pd.DataFrame(list(fan_out.items()), columns=["Signal", "Fan-out"])

# Save to CSV
df_fanin.to_csv("fan_in_data.csv", index=False)
df_fanout.to_csv("fan_out_data.csv", index=False)

# Print first few rows of the output
print("\nSample Fan-In Data:")
print(df_fanin.head())

print("\nSample Fan-Out Data:")
print(df_fanout.head())
```
